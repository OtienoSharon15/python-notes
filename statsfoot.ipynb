```python
import json
import pandas as pd

# Load the raw match data
with open("matches.json", "r") as f:
    data = json.load(f)

matches = data["matches"]
```

```python
# Extract relevant fields
match_list = []
for match in matches:
    if match['status'] == 'FINISHED':
        home = match['homeTeam']['name']
        away = match['awayTeam']['name']
        home_goals = match['score']['fullTime']['home']
        away_goals = match['score']['fullTime']['away']
        match_list.append({
            "home_team": home,
            "away_team": away,
            "home_goals": home_goals,
            "away_goals": away_goals
        })

# Save to CSV
df = pd.DataFrame(match_list)
df.to_csv("cleaned_matches.csv", index=False)
print("âœ… Cleaned data saved to cleaned_matches.csv")
```

```python
def get_winner(row): #to get winning teams
    if row["home_goals"] > row["away_goals"]:
        return row["home_team"]
    elif row["home_goals"] < row["away_goals"]:
        return row["away_team"]
    else:
        return "Draw"

df["winner"] = df.apply(get_winner, axis=1)
```

```python
df.head(10)
```

```python
from collections import defaultdict # getting the wins and appearances

team_counts = defaultdict(int)
win_counts = defaultdict(int)

for _, row in df.iterrows():
    home = row["home_team"]
    away = row["away_team"]
    winner = row["winner"]

    # Count appearances
    team_counts[home] += 1
    team_counts[away] += 1

    # Count wins (ignore draws)
    if winner != "Draw":
        win_counts[winner] += 1
```

```python
team_counts
win_counts
```

```python
import pandas as pd

teams = sorted(team_counts.keys())
summary = pd.DataFrame({
    "Team": teams,
    "Matches_Played": [team_counts[t] for t in teams],
    "Wins": [win_counts.get(t, 0) for t in teams]
})
summary["Win_Probability"] = (summary["Wins"] / summary["Matches_Played"]).round(3)

print(summary.sort_values(by="Win_Probability", ascending=False))
```
